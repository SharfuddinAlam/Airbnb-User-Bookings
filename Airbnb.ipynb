# Dates

Now,since the date has been converted into date timestamp,therefore it is possible to do data wrangling with Dates now.

```python
users.date_account_created.value_counts().plot()
sns.despine()
```

```python
users.date_first_active.value_counts().plot(kind='line',color='red')
plt.xlabel('year')
plt.ylabel('frequency')
sns.despine()
```

 The above two graphs shows a clear increase in the userbase of Airbnb in the time span of three years.Both the graph has similar patterns of growth.It is very much noticeable and this should give credits to Airbnb.However, the trend of growth is not smooth.There are peaks and hikes in October 2013 and right around April 2014.This gives a chance to do in depth analysis to find out what happens that may have caused such growth in the user base.

```python
 user_2013 = users[users.date_first_active > pd.to_datetime(20130101, format='%Y%m%d')]
user_2013=user_2013[user_2013.date_first_active < pd.to_datetime(20140101,format='%Y%m%d')]
user_2013.date_first_active.value_counts().plot(kind='line',linewidth=2.8,color='violet',fontsize=12)
sns.despine()
```

There is a steady pattern from the month of February till month of October.There is a sudden hike in October 2013.Let's see what's going on in there.

```python
user_October=users[(users.date_first_active > pd.to_datetime(20130930,format='%Y%m%d')) & (users.date_first_active <= pd.to_datetime(20131031,format='%Y%m%d'))]
user_October.date_first_active.value_counts().plot()
sns.despine()
```

At the first date of October 2013,which is a Tuesday,there were a lot of active users.Then the user activity dropped by significant amount in 4th of the month and kept on dropping throughout weekends.It again started to pick up on October 7th,which is a Tuesday and kept on rising at a steady pace.This clearly shows a trend that users are more tend to book in the middle of the weeks and active on weekdays rather than weekends.During weekends,there is a clear trend of less activity of booking which makes perfect sense.It may be that the users book during the weeks for the upcoming weekends for their vacation destination and tend to book less for the weekdays as they have to attend work.
